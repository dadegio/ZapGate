```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# =============================
# 1. Dati radar chart
# =============================
data = {
    "Criterio": ["Facilità d'uso", "Autonomia", "Trasparenza", "Velocità", "Privacy", "Setup"],
    "Custodial": [5, 1, 2, 5, 2, 5],
    "Non-Custodial": [2, 5, 5, 3, 5, 2]
}
df_radar = pd.DataFrame(data)

# Angoli per radar
labels = df_radar["Criterio"].tolist()
num_vars = len(labels)
angles = np.linspace(0, 2*np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

custodial = df_radar["Custodial"].tolist()
custodial += custodial[:1]
non_custodial = df_radar["Non-Custodial"].tolist()
non_custodial += non_custodial[:1]

fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))
ax.plot(angles, custodial, label="Custodial")
ax.fill(angles, custodial, alpha=0.25)
ax.plot(angles, non_custodial, label="Non-Custodial")
ax.fill(angles, non_custodial, alpha=0.25)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
ax.set_yticks([1,2,3,4,5])
ax.set_yticklabels(["1","2","3","4","5"])
plt.title("Confronto qualitativo Custodial vs Non-Custodial", pad=20)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Scenario": ["A - Giornalista", "B - Rivista", "C - Editore"],
    "Utenti": [1000, 10000, 100000],
    "Zap_medi": [5, 10, 20],
    "Valore_sats": [50, 100, 200],
    "Ricavo_euro": [100, 4000, 160000]
}
df = pd.DataFrame(data)

plt.plot(df["Utenti"], df["Ricavo_euro"], marker="o")
plt.title("Ricavi mensili per numero di utenti")
plt.xlabel("Utenti attivi")
plt.ylabel("Ricavi (€)")
plt.xscale("log")
plt.grid(True)
plt.show()
```

```python
import numpy as np

zap_range = np.arange(0, 500, 50)
valore_zap = 50  # sats
eur_per_sat = 0.0004  # tasso medio 1 sat ≈ 0.0004 €
spesa_zap = zap_range * valore_zap * eur_per_sat

plt.plot(zap_range, spesa_zap, label="Spesa Zap")
plt.axhline(y=5, color="r", linestyle="--", label="Abbonamento (5 €)")
plt.title("Break-even Zap vs Abbonamento")
plt.xlabel("Zap al mese")
plt.ylabel("Spesa utente (€)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
users = ["Casual", "Heavy"]
spesa_medio = [200, 7500]  # in sats
utenti_pct = [0.8, 0.2]

ricavi_tot = [spesa_medio[0]*800, spesa_medio[1]*200]  # esempio su 1000 utenti

df = pd.DataFrame({"Utenti": users, "Ricavi_sats": ricavi_tot})
df["Ricavi_cum_pct"] = df["Ricavi_sats"].cumsum() / df["Ricavi_sats"].sum() * 100

fig, ax1 = plt.subplots()

ax1.bar(df["Utenti"], df["Ricavi_sats"], color="skyblue")
ax1.set_ylabel("Ricavi totali (sats)")

ax2 = ax1.twinx()
ax2.plot(df["Utenti"], df["Ricavi_cum_pct"], color="red", marker="o", label="Cumulativa")
ax2.set_ylabel("% ricavi cumulativi")

plt.title("Distribuzione ricavi (Pareto 80/20)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Valori delle transazioni da 0.01€ a 5€
valori = np.linspace(0.01, 5, 100)

# Commissioni tipiche
paypal = 0.35 + valori*0.03
stripe = 0.25 + valori*0.029
zap = valori*0.0001  # praticamente trascurabile

plt.plot(valori, paypal, label="PayPal")
plt.plot(valori, stripe, label="Stripe")
plt.plot(valori, zap, label="Zap/LN")

plt.title("Confronto commissioni per transazione")
plt.xlabel("Valore transazione (€)")
plt.ylabel("Commissione (€)")
plt.legend()
plt.grid(True)
plt.show()
```
